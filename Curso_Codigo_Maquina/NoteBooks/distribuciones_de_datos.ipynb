# Distribuciones de probabilidad
## Normal, uniforme y exponencial

```python
normal = np.random.normal(size=10000000)
plt.hist(normal, color='pink', ec='black')
plt.title('Distribuci√≥n normal')
plt.show()

uniforme = np.random.uniform(size=10000000)
plt.hist(uniforme, color='green', ec='black')
plt.title("Distribuci√≥n uniforme")
plt.show()

expo = np.random.exponential(size=10000000)
plt.hist(expo, color='blue', ec='black')
plt.title('Distribuci√≥n exponencial')
plt.show()
```
