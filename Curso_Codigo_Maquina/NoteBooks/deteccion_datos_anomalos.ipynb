```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from time import perf_counter
```

```python
edades = np.array([4, 4, 3, 3, 5, 17, 29, 27, 50, 33, 61, 98, 56, 31, 92, 80, 18, 51, 33, 46])
edades = np.sort(edades)
edades, len(edades)/4
```

```python
plt.boxplot(edades)
plt.yticks(np.arange(0,110, 10))
plt.grid(True)
plt.show()
```

```python
np.min(edades), np.quantile(edades, 0.25), np.median(edades), np.quantile(edades, 0.75), np.max(edades)
```

```python
edades = np.array([10,
                  41, 51, 52, 54, 57, 60, 61, 65,
                  95, 100])
plt.boxplot(edades)
plt.yticks(np.arange(0, 110, 10))
plt.grid(True)
plt.show()
```

```python
print(f"Min: {np.min(edades)}\nCuartil 1: {np.quantile(edades, 0.25)}\
        \nMediana: {np.median(edades)}\nCuartil 3: {np.quantile(edades, 0.75)}\
        \nMax: {np.max(edades)}")
```
